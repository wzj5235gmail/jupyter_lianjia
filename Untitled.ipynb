```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# 读取数据库
con = sqlite3.connect(r'F:\Scrapy Projects\lianjiaSpider\lianjia_db.sqlite3')
cur = con.cursor()

# 读取2021-10-07爬取的数据，展示数据结构
data_1007 = cur.execute('SELECT * FROM Scraped20211007').fetchall()
columns = [i[1] for i in cur.execute('PRAGMA table_info(Scraped20211007)').fetchall()]
df_1007 = pd.DataFrame(data_1007, columns=columns)
df_1007.head(2)
```

```python
# 读取2021-12-02爬取的数据
data_1202 = cur.execute('SELECT * FROM Scraped20211202').fetchall()
columns = [i[1] for i in cur.execute('PRAGMA table_info(Scraped20211202)').fetchall()]
df_1202 = pd.DataFrame(data_1202, columns=columns)
df_1202.head(2)
```

```python
# 查看2021-10-07爬取的房源数量
df_1007.title.count()
```

```python
# 查看2021-12-02爬取的房源数量
df_1202.title.count()
```

```python
# 2021-10-07的数据，按不同城市，针对单价进行简单的描述统计
df_1007.groupby('city').unit_price.describe()
```

```python
# 2021-12-02的数据，按不同城市，针对单价进行简单的描述统计
df_1202.groupby('city').unit_price.describe()
```

```python
# 柱状图对比2021-10-07各城市平均单价
mean_1007 = df_1007.groupby('city').unit_price.mean()
x = mean_1007.index
plt.bar(x, mean_1007)
plt.title('Average Unit Price (2021-10-07)')
```

```python
# 柱状图对比2021-12-02各城市平均单价
mean_1202 = df_1202.groupby('city').unit_price.mean()
x = mean_1202.index
plt.bar(x, mean_1202, color='r')
plt.title('Average Unit Price (2021-12-02)')
```

```python
# 柱状图对比2021-10-07各城市中位数单价
median_1007 = df_1007.groupby('city').unit_price.median()
x = mean_1007.index
plt.bar(x, median_1007)
plt.title('Average Median Price (2021-10-07)')
```

```python
# 柱状图对比2021-12-02各城市中位数单价
median_1202 = df_1202.groupby('city').unit_price.median()
x = mean_1202.index
plt.bar(x, median_1202, color='r')
plt.title('Average Median Price (2021-12-02)')
```

```python
# 各城市近2个月平均单价变化幅度
mean_1202 / mean_1007 - 1
```

```python
# 各城市近2个月中位数单价变化幅度
median_1202 / median_1007 - 1
```

```python
# 2021-12-02北京各区域中位数面积、总价、单价统计，按单价从高到低排序
location_stats_bj_1202 = df_1202[df_1202['city']=='beijing'].groupby('location').median().sort_values('unit_price', ascending=False)
location_stats_bj_1202
```

```python
# 2021-12-02上海各区域中位数面积、总价、单价统计，按单价从高到低排序
location_stats_sh_1202 = df_1202[df_1202['city']=='shanghai'].groupby('location').median().sort_values('unit_price', ascending=False)
location_stats_sh_1202
```

```python
# 2021-12-02广州各区域中位数面积、总价、单价统计，按单价从高到低排序
location_stats_gz_1202 = df_1202[df_1202['city']=='guangzhou'].groupby('location').median().sort_values('unit_price', ascending=False)
location_stats_gz_1202
```

```python
# 2021-12-02深圳各区域中位数面积、总价、单价统计，按单价从高到低排序
location_stats_sz_1202 = df_1202[df_1202['city']=='shenzhen'].groupby('location').median().sort_values('unit_price', ascending=False)
location_stats_sz_1202
```

```python
# 对2021-10-07的数据重复一遍上述的统计
location_stats_bj_1007 = df_1007[df_1007['city']=='beijing'].groupby('location').median().sort_values('unit_price', ascending=False)
location_stats_sh_1007 = df_1007[df_1007['city']=='shanghai'].groupby('location').median().sort_values('unit_price', ascending=False)
location_stats_gz_1007 = df_1007[df_1007['city']=='guangzhou'].groupby('location').median().sort_values('unit_price', ascending=False)
location_stats_sz_1007 = df_1007[df_1007['city']=='shenzhen'].groupby('location').median().sort_values('unit_price', ascending=False)
```

```python
# 北京各区域近2个月面积、总价、单价涨幅，按单价涨幅从高到低排序
(location_stats_bj_1202 / location_stats_bj_1007 - 1).sort_values('unit_price', ascending=False).dropna()
```

```python
# 上海各区域近2个月面积、总价、单价涨幅，按单价涨幅从高到低排序
(location_stats_sh_1202 / location_stats_sh_1007 - 1).sort_values('unit_price', ascending=False).dropna()
```

```python
# 广州各区域近2个月面积、总价、单价涨幅，按单价涨幅从高到低排序
(location_stats_gz_1202 / location_stats_gz_1007 - 1).sort_values('unit_price', ascending=False).dropna()
```

```python
# 深圳各区域近2个月面积、总价、单价涨幅，按单价涨幅从高到低排序
(location_stats_sz_1202 / location_stats_sz_1007 - 1).sort_values('unit_price', ascending=False).dropna()
```

```python
# 计算每个房源近2月涨幅
df_1007.index = df_1007.link
df_1202.index = df_1202.link
df_1202['change'] = (df_1202.unit_price / df_1007.unit_price - 1)
df_1202.index = range(1, len(df_1202)+1)
df_change = df_1202[['city','title','name','location','unit_price','change','link']].dropna().sort_values('change', ascending=False)
```
